#### Load Test Data

```python
test_df = pd.read_csv("data/test.csv", encoding="utf-8")
test_df.head(5)
```

```python
test_df.dtypes
```

```python
test_df.drop(["Name", "Ticket", "Cabin", "PassengerId"], axis=1, inplace=True)
```

```python
test_df.head(5)
```

```python
test_df["Pclass"] = test_df["Pclass"].astype(str)
df_encoded = pd.get_dummies(test_df, prefix=["Pclass", "Sex", "Embarked"])
```

```python
# convert the new columns to int
col_lst = list(df_encoded.columns)
col_lst = col_lst[-8:]
for col in col_lst:
    df_encoded[col] = df_encoded[col].astype(int)
```
