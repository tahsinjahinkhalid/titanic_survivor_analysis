### Part III: Data Preparation for Model Selection

#### Load Essential Modules

```python
import numpy as np
import pandas as pd
import warnings
```

```python
# ignore deprecation/other warnings
warnings.filterwarnings('ignore')
```

#### Load Data

```python
titanic_df = pd.read_pickle("interim_data/preprocessed.pkl")
```

```python
titanic_df.drop("PassengerId", axis=1, inplace=True)
titanic_df.drop("Ticket", axis=1, inplace=True)
titanic_df.drop("Name", axis=1, inplace=True)
```

```python
# this is a categorical according to the Kaggle data card
titanic_df["Pclass"] = titanic_df["Pclass"].astype(str)
```

```python
titanic_df.head(5)
```

```python
df_encoded = pd.get_dummies(titanic_df, prefix=["Pclass", "Sex", "Embarked"])
```

```python
df_encoded.head(10)
```

```python
# convert the new columns to int
col_lst = list(df_encoded.columns)
col_lst = col_lst[-8:]
for col in col_lst:
    df_encoded[col] = df_encoded[col].astype(int)
```

```python
df_encoded.head(10)
```

#### Summary of Notebook

We have dummy encoded the three categorical columns and the dataset is now ready to be used in Model selection

#### Make Pickle

```python
df_encoded.to_pickle("interim_data/ready_ML.pkl")
```
